```python
import matplotlib
import matplotlib.pyplot as plt
import random
import math
```

```python
class StatChart:
    def __init__(self, foo_name):
        self.foo_name = foo_name
        
        
    def get_coordinates(self):
        timeit = Timer()
        self.lens = [10, 100, 1000, 10000, 100000, 1000000]
        self.times = []
        for len_array in self.lens:
            self.array = random.sample(range(len_array + 42), len_array)
            with timeit:
                self.foo_name(self.array, 42)
            self.times.append(timeit.delta.total_seconds())
        return self.lens, self.times
    
    
    def get_chart(self):
        plt.plot(self.lens, self.times, 'bo-') 
        return 
```

# Simple search
Complexity: $O(n)$

```python
from searches import dummy_search
from timer_context_manager import Timer
```

```python
simple = StatChart(dummy_search)
simple_lens, simple_times = simple.get_coordinates()
simple.get_chart()
```

# Binary search

Complexity: $O(log n)$

```python
from searches import binary_search
```

```python
binary = StatChart(binary_search)
binary_lens, binary_times = binary.get_coordinates()
binary.get_chart()
```

# Comparision 
Simple vs binary search

```python
plt.plot(simple_lens, simple_times, 'bo-', label='Simple search')
plt.plot(binary_lens, binary_times, 'go-', label='Binary search')
plt.plot(simple_lens, simple_lens, 'ro-', label='Linear function')
plt.plot(binary_lens, [math.log2(x) for x in binary_lens], 
         'yo-', label='Log function')
plt.legend(loc='best', ncol=2, shadow=True, fancybox=True)
```
