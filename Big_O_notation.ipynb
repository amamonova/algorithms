```python
import matplotlib
import matplotlib.pyplot as plt
import random
import math
```

```python
class StatChart:
    def __init__(self, foo_name):
        self.foo_name = foo_name
        
        
    def get_coordinates(self):
        timeit = Timer()
        times = []
        lens = [10, 100, 1000, 10000, 100000, 1000000]
        
        for len_array in lens:
            array = random.sample(range(len_array + 42), len_array)
            with timeit:
                self.foo_name(array, 42)
            times.append(timeit.delta.total_seconds())
        return lens, times
    
    
    def get_chart(self, lens, times):
        plt.plot(lens, times, 'bo-') 
        return 
```

# Simple search
Complexity: $O(n)$

```python
from searches import dummy_search
from timer_context_manager import Timer
```

```python
simple = StatChart(dummy_search)
simple_lens, simple_times = simple.get_coordinates()
simple.get_chart(simple_lens, simple_times)
```

# Binary search

Complexity: $O(log n)$

```python
from searches import binary_search
```

```python
binary = StatChart(binary_search)
binary_lens, binary_times = binary.get_coordinates()
binary.get_chart(binary_lens, binary_times)
```

# Comparision 
Simple vs binary search

```python
plt.plot(simple_lens, simple_times, 'bo-', label='Simple search')
plt.plot(simple_times, simple_times, 'ro-', label='Linear function')
plt.legend(loc='best', ncol=2, shadow=True, fancybox=True)
```

```python
plt.plot(binary_lens, binary_times, 'go-', label='Binary search')
plt.plot(binary_lens, [math.log2(x) for x in binary_lens], 
         'yo-', label='Log function')
plt.legend(loc='best', ncol=2, shadow=True, fancybox=True)
```
